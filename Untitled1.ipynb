```python
import google.generativeai as genai
from getpass import getpass
```

```python
api_key = getpass("enter your api key:")
genai.configure(api_key=api_key)
```

```python
class GeminiAgent:
    def __init__(self, name, role, model="gemini-2.0-flash"):
        self.name = name
        self.role = role
        self.model = genai.GenerativeModel(model)
        self.chat = self.model.start_chat(history=[])
        self.terminated = False
    def process(self, message):
        if self.terminated:
            return f"{self.name} has ended its participation."

        prompt = (
            f"You are {self.name}, a DSA AI agent specializing as {self.role}.\n"
            f"Analyze or respond to the following message:\n\n{message}"
        )

        response = self.chat.send_message(prompt).text.strip()

        # Termination strategy
        if any(x in response.lower() for x in ["no further", "nothing more", "no additional issues", "no comments"]):
            self.terminated = True

        return response
```

```python
class ChatOrchestrator:
    def __init__(self, agents):
        self.agents = agents

    def initiate_chat(self, user_message, max_rounds=5):
        print(f"üßë User: {user_message}\n")
        message = user_message

        for round_num in range(max_rounds):
            print(f"üîÅ Round {round_num + 1}")
            all_terminated = True

            for agent in self.agents:
                if not agent.terminated:
                    reply = agent.process(message)
                    print(f"{agent.name}: {reply}\n")
                    message = reply  # update message for next agent
                    all_terminated = False

            if all_terminated:
                print("‚úÖ Agents have determined the conversation is complete.")
                break
```

```python
dsa_mentor = GeminiAgent(
    name="DSA Mentor",
    role="expert in explaining algorithms and complexity analysis"
)

code_optimizer = GeminiAgent(
    name="Code Optimizer",
    role="specialist in improving time and space efficiency of solutions"
)

edge_case_hunter = GeminiAgent(
    name="Edge Case Hunter",
    role="expert at identifying boundary and failing test cases"
)

orchestrator = ChatOrchestrator([
    dsa_mentor,
    code_optimizer,
    edge_case_hunter
])
```

```python
problem ="""
Can you explain how to solve the palindrome problem in Python? \
    Show both iterative and recursive approaches, and explain edge cases.
"""


orchestrator.initiate_chat(problem)
```
